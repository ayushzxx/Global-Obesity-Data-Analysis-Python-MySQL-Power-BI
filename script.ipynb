```python
import pandas as pd

df = pd.read_csv("data.csv", header=None)

rows = []

# Loop through each country first
for r in range(4, len(df)):

    country = df.iloc[r, 0]

    # Then each year
    for col in range(1, df.shape[1], 3):

        year = int(df.iloc[0, col])

        # Then each sex
        for i in range(3):

            sex = df.iloc[3, col + i]

            obesity = str(df.iloc[r, col + i]).split(" ")[0]
            obesity = pd.to_numeric(obesity, errors="coerce")

            rows.append({
                "Country": country,
                "Year": year,
                "Sex": sex,
                "Obesity %": obesity
            })

final_df = pd.DataFrame(rows)

final_df.head(12)
```
